```python
#Matplotlib is a plotting library for the Python programming language
#Matplotlib makes Data visualization quick and easy by plotting the data in graph manner.
```

```python
#importing matplot lib
import matplotlib.pyplot as plt
```

```python
#Write a python program to draw a line using given axis values with suitable label in the x-axis and y-axis and the title
x = [1,2,3]
y = [2,3,1]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple graph')
plt.show()
```

```python
#Write a python program to draw two or more lines with different styles
x2 = [5,3,1]
y2 = [3,2,5]
plt.plot(x,y,color = 'blue',linewidth = 3, label = 'line1-dotted',linestyle = 'dotted')
plt.plot(x2,y2,color = 'red', linewidth = 5, label = 'line2-dashed',linestyle = 'dashed')
plt.title("Two lines with different styles")
plt.show()
```

```python
#Write a python program to display the bar chart of the popularity of the programming languages
languages = ['Java','Python','PHP','JavaScript','C#','C++']
popularity = [22.2,17,8,9,7,6]
plt.bar(languages,popularity,color='green')
plt.grid()
#grid90 will display grids within the graph
plt.show()
```

```python
#Write a python program to create bar plot of scores by group and gender
#Use multiple X values on the ame chart for men and women
import numpy as np

men = (22,30,33,30,26)
women = (25,32,30,35,29)
groups =5

index = np.arange(5)
bar_width = 0.35
opacity = 0.8

r1 = plt.bar(index,men,bar_width,alpha = opacity)
r2 = plt.bar(index + bar_width, women,bar_width,alpha = opacity)

plt.xlabel("Men")
plt.ylabel("Women")
plt.title("Bar graph")
plt.xticks(index + bar_width,('G1','G2','G3','G4','G5'))

plt.show()
```

```python
#Write a python program to display horizontal barchart of popular programming languages
plt.barh(languages,popularity,color = 'green')
#Plotting major and Minor grids (Optional)
plt.minorticks_on()
plt.grid(which = 'major',linestyle = '-',color = 'red')
plt.grid(which = 'minor',linestyle = ':',color = 'green')
```
