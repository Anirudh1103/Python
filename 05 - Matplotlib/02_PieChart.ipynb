```python
import matplotlib.pyplot as plt
```

```python
#Write a python program to create a pie chart of the popularity of the programming languages

languages = ['Java','Python','PHP','JavaScript','C#','C++']
popularity = [22.2,17,8,9,7,6]
color = ["blue","red","orange","purple","brown","green"]
explode = (0.1,0,0,0,0,0)

plt.pie(popularity,labels = languages, colors = color,shadow = True,explode = explode,startangle = 140)

plt.show()
```

```python
#Write a Python program to create a pie chart with titleof popularity of programming languages
plt.pie(popularity,labels = languages, colors = color,shadow = True,autopct = '%1.1f%%',explode = explode,startangle = 140)
"""
    Explaination of the attribute in the above code within pie()
    popularity : an list containing data of popularity of programming languages
    labels = languages: languages is a list containing names of programming languages we are passing 
                        it to the labels attribute
    colors = color: color is list containing names of colors we are passing it to the colors attribute
    shadow = True : This will add shadow to the pie chart
    autopct = '%1.1f%%' : This will specify the percentage within the pie chart
    explode = explode : This will explode the first slice
    for exploding the second slice the data would be (0,0.1,0,0,0,0)
    startangle = 140 : This will specify the start angle or the first piece of the pie chart.
"""
plt.title("Popularity of the programming languages")
plt.show()
```
